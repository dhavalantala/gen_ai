```python
import string

letter = string.ascii_lowercase


for i in range(len(letter)):

    if i == 0 or i == 20:
        ordind = "st"
    elif i == 1 or i == 21:
        ordind = "nd"
    elif i == 2 or i == 22:
        ordind = "rd"
    else:
        ordind = 'th'
    # print("%s is the %g%s letter of the alphabat"%(letter[i], i+1, ordind))
    print(f"{letter[i]} is the {i+1}{ordind} letter of the alphabate")
    
```

```python
x = np.linspace(-3, 3, 101)

plt.plot(x,x, label = "y==x")
plt.plot(x, x**2, label = "y==x**2")
plt.plot(x, x**3, label = "y==x**3")

plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.xlim([x[0],x[-1]])
plt.ylim([-10, 10])
plt.title("A really awesome plot!")

# plt.axis('equal')
plt.gca().set_aspect(.3)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

plt.plot(x,x,'r',  label = "y==x")
plt.plot(x, x**2, label = "y==x**2")
plt.plot(x, x**3, label = "y==x**3", color = [0.8, 0.1, 0.7])

plt.legend()

ax.set_xlabel('x')
ax.set_ylabel('y=f(x)')
ax.set_title("A really awesome plot!")
ax.set_xlim([x[0], x[-1]])
ax.set_ylim([-10, 10])

ax.set_aspect(1/ax.get_data_ratio())

ax.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

for i in np.linspace(0, 1, 50):
    y = x**i
    plt.plot(x, y, color=[i/2, 0, i])  # Smooth colorful curves

plt.title("Multiple Lines: y = x^i")
plt.show()
```

```python
x = np.arange(-5, 6)
y = x**2

fig, ax = plt.subplots()
ax.plot(x,y, 'ko-', markerfacecolor='w', markersize = 15)
ax.annotate('Hello', (x[3], y[3]), fontsize = 15, arrowprops=dict(), xycoords = 'data', xytext = (0, 10), horizontalalignment= 'center')

plt.show()
```

```python
x = np.arange(-5, 6)
y = x**2 + np.cos(x) * 10

minpont = np.argmin(y)  ## It give you a index value not min value
txt = 'min : (%g, %.2f)'%(x[minpont], y[minpont])

fig, ax = plt.subplots()
ax.plot(x,y, 'ko-', markerfacecolor='purple', markersize = 15)
ax.annotate(txt, (x[2], y[2]), fontsize = 15, 
            xycoords = 'data', xytext = (x[int(len(x)/2)], np.max(y)), horizontalalignment= 'center',
            arrowprops=dict(color = 'purple', arrowstyle = 'wedge'))

ax.set_xlabel("x")
ax.set_ylabel("f(x) = x^2 + 10cos(x)")

plt.show()
```

```python
n = 200 
D = np.zeros((n,2))

D[:, 0] = np.linspace(0, 10, n) + np.random.randn(n)
D[:, 1] = D[:, 0]**2 + np.random.randn(n)*10


# sns.jointplot((D[:, 0], D[:, 1]))
sns.jointplot(x=D[:, 0], y=D[:, 1])
plt.show()
```

```python
import pandas as pd
```

```python
df  = pd.DataFrame(data=D, columns=['var1', 'var2'])

sns.jointplot(data=df, x=df.columns[0], y=df.columns[1], kind='kde', color='purple')
plt.show()
```

```python
n = 10
H = np.zeros((n,n))

for i in range(1, n+1):
    for j in range(1, n+1):
        H[i-1, j-1] = 1 / (i+j-1)

np.round(H, 2)

sns.heatmap(H, vmin = 0, vmax = 0.3)
plt.show()
```
